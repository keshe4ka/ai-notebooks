#Обработка данных#

Заполняем пропуски

```python
data['MinTemp'] = data['MinTemp'].fillna(10.2)

data['MaxTemp'] = data['MaxTemp'].fillna(20.0)

data['Rainfall'] = data['Rainfall'].fillna(0.0)

data['Evaporation'] = data['Evaporation'].fillna(4.0)

data['Sunshine'] = data['Sunshine'].fillna(0.0)

data['WindGustDir'] = data['WindGustDir'].fillna('W')

data['WindGustSpeed'] = data['WindGustSpeed'].fillna(35.0)

data['WindDir9am'] = data['WindDir9am'].fillna('N')

data['WindDir3pm'] = data['WindDir3pm'].fillna('SE')

data['WindSpeed9am'] = data['WindSpeed9am'].fillna(9.0)

data['WindSpeed3pm'] = data['WindSpeed3pm'].fillna(17.0)

data['Humidity9am'] = data['Humidity9am'].fillna(99.0)

data['Humidity3pm'] = data['Humidity3pm'].fillna(52.0)

data['Pressure9am'] = data['Pressure9am'].fillna(1016.0)

data['Pressure3pm'] = data['Pressure3pm'].fillna(1015.5)

data['Cloud9am'] = data['Cloud9am'].fillna(7.0)

data['Cloud3pm'] = data['Cloud3pm'].fillna(7.0)

data['Temp9am'] = data['Temp9am'].fillna(17.0)

data['Temp3pm'] = data['Temp3pm'].fillna(20.0)
```

```python
data.info()
```

Бинаризируем данные

```python
data = pd.get_dummies(data, categorial_columns)
data.head()
```

#Строим модель#

Поделим данные на признаки и целевую переменную

```python
y = data['RainTomorrow']
X = data.drop('RainTomorrow', axis=1)
```

```python
from sklearn import ensemble #, classifier_model
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
rf = ensemble.RandomForestClassifier(n_estimators=100, max_depth=10, min_samples_split=7)
```

```python
rf.fit(X,y)
```
